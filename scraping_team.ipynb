```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd

# Initialize Selenium
chrome_service = ChromeService("C:/Users/ALESSANDRO/Downloads/chromedriver.exe")
chrome_service.start()
chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

# URL of the webpage
url = "https://www.daviscup.com/en/draws-results/tie.aspx?id=M-DC-2018-WG-M-FRA-NED-01"

# Navigate to the webpage
driver.get(url)

try:
    # Find the div element with class "component-title"
    component_title_element = driver.find_element(By.CSS_SELECTOR, "div.component-title")
    component_title_text = component_title_element.text.strip()

    # Find the div element with class "tie"
    tie_element = driver.find_element(By.CSS_SELECTOR, "div.tie")
    tie_text = tie_element.text.strip()
    tie_text_parts = tie_text.split(":", 1)
    variable_name = tie_text_parts[0].strip()
    variable_content_parts = tie_text_parts[1].strip().split("\n")
    date = variable_content_parts[0].strip()
    venue = variable_content_parts[1].strip()

    # Find all div elements with class "rubber-header"
    rubber_header_elements = driver.find_elements(By.CSS_SELECTOR, "div.rubber-header")

    # Find all elements with class "rubber-body"
    rubber_body_elements = driver.find_elements(By.CLASS_NAME, "rubber-body")

    # Create lists to store the extracted data
    stages = []
    variables = []
    matches = []
    statuses = []  # New list for statuses
    players = []
    scores = []

    for rubber_header_element, rubber_body_element in zip(rubber_header_elements, rubber_body_elements):
        stage = component_title_text  # Use the component title as the stage for all rows
        variable_name = rubber_header_element.text.strip()
        
        # Assign the variable name and content to the corresponding lists
        stages.append(stage)
        variables.append(variable_name)

        # Split the variable into match and status
        variable_parts = variable_name.split("\n")
        if len(variable_parts) >= 2:
            match = variable_parts[0].strip()
            status = variable_parts[1].strip()
        else:
            match = variable_name.strip()
            status = None

        matches.append(match)
        statuses.append(status)  # Add status to the list

        # Assign the rubber header text to the following rubber body as the "match"
        match = rubber_body_element.text.strip()
        matches.append(match)

        # Split the information in the rubber-body according to the two cells
        rows = rubber_body_element.find_elements(By.TAG_NAME, "tr")
        for row in rows:
            cells = row.find_elements(By.TAG_NAME, "td")
            if len(cells) >= 2:
                player = cells[0].text.strip()
                score = cells[1].text.strip()
                players.append(player)
                scores.append(score)

    # Make sure all lists have the same length
    min_length = min(len(stages), len(variables), len(matches), len(statuses), len(players), len(scores))
    stages = stages[:min_length]
    variables = variables[:min_length]
    matches = matches[:min_length]
    statuses = statuses[:min_length]
    players = players[:min_length]
    scores = scores[:min_length]

    # Create a DataFrame from the collected data
    data = {
        "Stage": stages,
        "Date": [date] * min_length,  # Use the same date for all rows
        "Venue": [venue] * min_length,  # Use the same venue for all rows
        "Variable": variables,
        "Match": matches,
        "Status": statuses,  # Add the status column
        "Player": players,
        "Score": scores
    }

    df = pd.DataFrame(data)

    # Display the DataFrame
    print(df)

except Exception as e:
    print("Error:", str(e))
finally:
    # Close the Selenium WebDriver
    driver.quit()
```
